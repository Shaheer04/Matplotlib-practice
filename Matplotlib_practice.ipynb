##Matplotlib
Matplotlib is a data visualization library for python and its the most used data visualization library in python.

```python
# import numpy and matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
X_data = np.random.random(1000) * 100
Y_data = np.random.random(1000) * 100
```

```python
# scatter plot
plt.scatter(X_data,Y_data, c="#000", marker="o", s=100, alpha=0.3)
plt.show()
```

```python
years = [2006 + x for x in range(10)]
weights = [100,105,90,97,96,89,88,82,89,93]
print(years)
print(weights)
```

```python
# line plot
plt.plot(years,weights, c='b', lw=2, linestyle= '--')
plt.show()
```

```python
x = ["c++", "python", "java", "rust", "GO", "javascript"]
y = [20,100,67,10,23,56]
```

```python
# Bar plot
plt.bar(x,y,width=0.5, color="y")
plt.plot()
```

```python
ages = np.random.normal(22,2,1000)
print(ages)
```

```python
plt.hist(ages, bins=15 , cumulative=True)
plt.show()
```

```python
langs = ["Python", "C++", "java", "GO"]
votes = [20,50,14,7]
explodes = [0,0,0,0.2]
```

```python
plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%')
plt.show()
```

```python
heights = np.random.normal(150,5,300)
print(heights)
```

```python
plt.boxplot(heights)
plt.show()
```

```python
stock_a = [67,78,45,89,45,77]
stock_b = [88,98,34,40,56,70]
stock_c = [60,64,70,88,82,90]

plt.plot(stock_a, label="c1")
plt.plot(stock_b, label="c2")
plt.plot(stock_c, label="c3")
plt.legend(loc="lower right")
plt.show()
```

```python
style.use("ggplot")
votes = [10,2,20,11]
people = ['A','B','C','D']

plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

```python
# plotting multiple plots at a time
x1,y1 = np.random.random(100), np.random.random(100)
x2,y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2,y2)
plt.show()
```

```python
# subplots in one window
x = np.arange(100)

fig, axis = plt.subplots(2, 2)

axis[0,0].plot(x,np.sin(x))
axis[0,0].set_title("Sine wave")

axis[0,1].plot(x,np.cos(x))
axis[0,1].set_title("Cosine wave")

axis[1,0].plot(x,np.random.random(100))
axis[1,0].set_title("random function")

axis[1,1].plot(x,np.log(x))
axis[1,1].set_title("Log function")
fig.suptitle("Four PLots")
plt.tight_layout() # it make sure that nothing overlaps
# importing graphic ploting
plt.savefig("fourplots.png", dpi=300)
```

```python
# 3d plots

ax = plt.axes(projection="3d")

x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
ax.set_title("3D plot")
plt.show()
```

```python
# Surface plot in 3d

ax = plt.axes(projection="3d")

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X,Y,Z, cmap="Spectral")
ax.set_title("Surface plot")
plt.show()
```
